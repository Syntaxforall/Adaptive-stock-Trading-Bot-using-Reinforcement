```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta

# Load dataset
data = pd.read_csv('/content/data1.csv', parse_dates=True, index_col='Date')

# Ensure necessary columns exist
if {'Close', 'Open', 'High', 'Low', 'Volume'}.issubset(data.columns):
    # Calculate Moving Averages
    data['SMA_50'] = ta.trend.sma_indicator(data['Close'], window=50)
    data['SMA_200'] = ta.trend.sma_indicator(data['Close'], window=200)

    # Calculate RSI
    data['RSI'] = ta.momentum.rsi(data['Close'], window=14)

    # Calculate MACD
    data['MACD'] = ta.trend.macd(data['Close'])
    data['Signal_Line'] = ta.trend.macd_signal(data['Close'])

    # Rule-based Trading Strategy
    def generate_signals(data):
        signals = []
        for i in range(len(data)):
            if data['SMA_50'][i] > data['SMA_200'][i] and data['RSI'][i] < 30 and data['MACD'][i] > data['Signal_Line'][i]:
                signals.append('BUY')
            elif data['SMA_50'][i] < data['SMA_200'][i] and data['RSI'][i] > 70 and data['MACD'][i] < data['Signal_Line'][i]:
                signals.append('SELL')
            else:
                signals.append('HOLD')
        return signals

    data['Signal'] = generate_signals(data)

    # Plotting the strategy
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Close Price', alpha=0.5)
    plt.plot(data.index, data['SMA_50'], label='SMA 50', linestyle='--')
    plt.plot(data.index, data['SMA_200'], label='SMA 200', linestyle='--')
    plt.scatter(data.index[data['Signal'] == 'BUY'], data['Close'][data['Signal'] == 'BUY'], marker='^', color='g', label='Buy Signal')
    plt.scatter(data.index[data['Signal'] == 'SELL'], data['Close'][data['Signal'] == 'SELL'], marker='v', color='r', label='Sell Signal')
    plt.legend()
    plt.title('Rule-Based Trading Strategy')
    plt.show()

    # Save results
    data.to_csv('/content/trading_signals.csv')
else:
    print("Error: Required columns are missing from the dataset.")
```

```python
import numpy as np
import pandas as pd
import gym
from gym import spaces
from stable_baselines3 import PPO
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/data.csv")  # Replace with your actual data path
data.columns = data.columns.str.strip()
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data.set_index('Date', inplace=True)
data = data[['Open', 'High', 'Low', 'Close', 'Volume']]

# Optimized preprocessing function
def preprocess_data(data, lookback=50):
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data)
    indices = np.arange(lookback, len(scaled_data))
    X = np.array([scaled_data[i-lookback:i] for i in indices])
    y = scaled_data[lookback:, 3]  # 'Close' price
    return X, y, scaler

# Preprocess data
X, y, scaler = preprocess_data(data)

# Optimized GRU model
def build_gru_model(input_shape):
    model = Sequential([
        GRU(32, return_sequences=False, input_shape=input_shape),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

# Train GRU model
gru_model = build_gru_model((X.shape[1], X.shape[2]))
gru_model.fit(X, y, epochs=5, batch_size=64, verbose=0, validation_split=0.2)

# Optimized Stock Trading Environment
class StockTradingEnv(gym.Env):
    def __init__(self, data, gru_model, lookback=50):
        super(StockTradingEnv, self).__init__()
        self.data = data.values
        self.gru_model = gru_model
        self.lookback = lookback
        self.current_step = lookback
        self.balance = 10000
        self.shares_held = 0
        self.net_worth = self.balance
        self.action_space = spaces.Discrete(3)
        self.observation_space = spaces.Box(low=0, high=1, shape=(lookback * data.shape[1] + 1,), dtype=np.float32)
        self.gru_predictions = self._precompute_gru_predictions()

    def _precompute_gru_predictions(self):
        historical_data = np.array([self.data[i-self.lookback:i] for i in range(self.lookback, len(self.data))])
        return self.gru_model.predict(historical_data, batch_size=64, verbose=0)

    def reset(self):
        self.current_step = self.lookback
        self.balance = 10000
        self.shares_held = 0
        self.net_worth = self.balance
        return self._get_observation()

    def step(self, action):
        current_price = self.data[self.current_step, 3]
        if action == 1:  # Buy
            self.shares_held += self.balance / current_price
            self.balance = 0
        elif action == 2:  # Sell
            self.balance += self.shares_held * current_price
            self.shares_held = 0
        self.net_worth = self.balance + self.shares_held * current_price
        reward = self.net_worth - 10000
        self.current_step += 1
        done = self.current_step >= len(self.data) - 1
        return self._get_observation(), reward, done, {}

    def _get_observation(self):
        historical_data = self.data[self.current_step - self.lookback:self.current_step].flatten()
        gru_pred = self.gru_predictions[self.current_step - self.lookback]
        return np.concatenate([historical_data, gru_pred])

# Train PPO model
env = StockTradingEnv(data, gru_model)
model = PPO("MlpPolicy", env, verbose=0)
model.learn(total_timesteps=5000)

# Testing
obs = env.reset()
rewards, net_worths, actions = [], [], []  # Added actions list
max_steps = len(data) - env.lookback
for _ in range(max_steps):
    action, _ = model.predict(obs, deterministic=True)
    actions.append(action)  # Collect action
    obs, reward, done, _ = env.step(action)
    rewards.append(reward)
    net_worths.append(env.net_worth)
    if done:
        break

# Confusion Matrix and F1 Score
net_worths = np.array(net_worths)
actions = np.array(actions)

# Infer true actions based on net worth changes (hindsight optimal)
trade_returns = np.diff(net_worths)
true_actions = np.zeros(len(actions), dtype=int)  # Default to hold (0)
true_actions[1:][trade_returns > 0] = 1  # Buy if net worth increases
true_actions[1:][trade_returns < 0] = 2  # Sell if net worth decreases

# Ensure lengths match
true_actions = true_actions[:len(actions)]

# Calculate confusion matrix
cm = confusion_matrix(true_actions, actions)

# Calculate F1 score (macro average for multi-class)
f1 = f1_score(true_actions, actions, average='macro')

# Plot confusion matrix
action_labels = ['Hold', 'Buy', 'Sell']
plt.figure(figsize=(6, 6))
plt.imshow(cm, interpolation='nearest', cmap='Blues')
plt.title(f'Confusion Matrix (F1 Score: {f1:.4f})')
plt.colorbar(label='Number of Instances')

tick_marks = np.arange(len(action_labels))
plt.xticks(tick_marks, action_labels, rotation=45)
plt.yticks(tick_marks, action_labels)

# Add text annotations
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, f'{cm[i, j]}',
                 ha='center', va='center',
                 color='white' if cm[i, j] > thresh else 'black')

plt.xlabel('Predicted Action')
plt.ylabel('True Action (Hindsight)')
plt.tight_layout()
plt.show()

# Print F1 score
print(f"F1 Score (Macro): {f1:.4f}")

# Generate sample data for total rewards and profits (for demonstration)
np.random.seed(42)
steps = 60
total_rewards = np.cumsum(np.random.uniform(-1, 2, steps))
total_profit = 1 + np.cumsum(np.random.uniform(-0.005, 0.01, steps))
reward_points = total_rewards[1:] - total_rewards[:-1] > 0
penalty_points = ~reward_points

# Plot the total reward graph
plt.figure(figsize=(6, 4))
plt.plot(total_rewards, label="Total Reward", color="blue")
plt.scatter(range(1, steps), total_rewards[1:], c=np.where(reward_points, "green", "red"))
plt.title(f"Total Reward: {total_rewards[-1]:.4f}")
plt.xlabel("Steps")
plt.ylabel("Reward")
plt.legend()
plt.show()

# Plot the total profit graph
plt.figure(figsize=(6, 4))
plt.plot(total_profit, label="Total Profit", color="blue")
plt.scatter(range(1, steps), total_profit[1:], c=np.where(reward_points, "green", "red"))
plt.title(f"Total Profit: {total_profit[-1]:.4f}")
plt.xlabel("Steps")
plt.ylabel("Profit")
plt.legend()
plt.show()

# Optimized plotting of actual results
plt.figure(figsize=(10, 5))
plt.plot(rewards[::10], label='Rewards')
plt.title('Rewards over Time')
plt.xlabel('Steps (downsampled)')
plt.ylabel('Reward')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(net_worths[::10], label='Net Worth')
plt.title('Net Worth over Time')
plt.xlabel('Steps (downsampled)')
plt.ylabel('Net Worth')
plt.legend()
plt.show()

print(f"Final Net Worth: {env.net_worth}")
```

```python
import numpy as np

def print_trading_metrics(net_worths):
    """
    Calculate and print trading performance metrics based on net worths.

    Parameters:
    - net_worths: List or array of net worth values over time from the trading simulation.
    """
    # Convert to numpy array if not already
    net_worths = np.array(net_worths)

    # Max Drawdown Calculation
    peak = np.maximum.accumulate(net_worths)
    drawdown = (peak - net_worths) / peak
    max_drawdown = np.max(drawdown) * 100  # Convert to percentage

    # Trade Profit/Loss Calculation
    trade_returns = np.diff(net_worths)  # Profit/Loss per trade
    wins = trade_returns[trade_returns > 0]
    losses = trade_returns[trade_returns < 0]

    # Win Rate
    total_trades = len(trade_returns)
    win_rate = len(wins) / total_trades if total_trades > 0 else 0

    # Average Profit/Loss per Trade
    avg_profit_loss = np.mean(trade_returns) if total_trades > 0 else 0

    # Profit Factor (Total Profit / Total Loss)
    total_profit = np.sum(wins)
    total_loss = np.abs(np.sum(losses))
    profit_factor = total_profit / total_loss if total_loss > 0 else np.inf

    # Sharpe Ratio
    risk_free_rate = 0.01  # Assume 1% annual risk-free return
    excess_returns = trade_returns - risk_free_rate / 252  # Daily excess returns
    std_excess_returns = np.std(excess_returns)
    sharpe_ratio = np.mean(excess_returns) / std_excess_returns if std_excess_returns > 1e-6 else 0

    # Print all metrics
    print("Trading Performance Metrics:")
    print(f"Max Drawdown: {max_drawdown:.2f}%")
    print(f"Win Rate: {win_rate * 100:.2f}%")
    print(f"Average Profit/Loss per Trade: {avg_profit_loss:.4f}")
    print(f"Profit Factor: {profit_factor:.4f}")
    print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
```
